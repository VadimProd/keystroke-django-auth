```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 1. Структура данных

```python
DATASET_PATH = '/content/DSL-StrongPasswordData.csv'

df = pd.read_csv(DATASET_PATH)
subject = df['subject']
df.head()
```

# 2. Распределение числа записей по пользователям

```python
import seaborn as sns

plt.figure(figsize=(12, 5))
sns.countplot(x='subject', data=df, palette='coolwarm')
plt.title('Количество записей по пользователям')
plt.xticks(rotation=90)
plt.xlabel('ID пользователя')
plt.ylabel('Количество попыток')
plt.tight_layout()
plt.show()
```

# 3. Пример поведенческого профиля

```python
user_df = df[df['subject'] == 's004']  # Пример пользователя
hold_columns = [col for col in df.columns if col.startswith('H.')]

plt.figure(figsize=(10, 5))
user_df[hold_columns].iloc[0:50].T.plot(kind='line', legend=False)
plt.title('Профиль удержания клавиш (Hold Time) для пользователя s004 (50 первых попыток)')
plt.xlabel('Клавиши')
plt.ylabel('Время удержания (сек)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
